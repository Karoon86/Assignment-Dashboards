```python


# Import the required libraries
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

# Load the data into a pandas DataFrame
df = pd.read_csv('m5_survey_data_technologies_normalised.csv')

# Create the dashboard layout
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Top 5 Languages", "Language Popularity")
)

# Add the first chart (Top 5 Languages)
top_languages = df['LanguageWorkedWith'].value_counts().head(5)
fig.add_trace(
    go.Bar(x=top_languages.index, y=top_languages.values),
    row=1, col=1
)




# Add the fourth chart (Language Popularity)
language_popularity = df.groupby('LanguageWorkedWith')['LanguageWorkedWith'].count().sort_values(ascending=False)
fig.add_trace(
    go.Bar(x=language_popularity.index, y=language_popularity.values),
    row=2, col=2
)

# Customize the layout
fig.update_layout(
    title='LanguageWorkedWith Dashboard',
    height=600,
    width=900,
    showlegend=False
)

# Display the dashboard
fig.show()
```

Visualize as a Hierarchy bubble chart.

```python
import numpy as np
import pandas as pd
from scipy import ndimage
from scipy.cluster import hierarchy
from scipy.spatial import distance_matrix
from matplotlib import pyplot as plt
from sklearn import manifold, datasets
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
```

Generating Random Data

We will be generating a set of data using the make_blobs class.

```python
X1, y1 = make_blobs(n_samples=50, centers=[[4,4], [-2, -1], [1, 1], [10,4]], cluster_std=0.9)
```

Plot the scatter plot of the randomly generated data.

```python
plt.scatter(X1[:, 0], X1[:, 1], marker='o')
```

Agglomerative Clustering

We will start by clustering the random data points we just created.

```python
agglom = AgglomerativeClustering(n_clusters = 4, linkage = 'average')
```

```python
agglom.fit(X1,y1)
```

Run the following code to show the clustering!

Remember to read the code and comments to gain more understanding on how the plotting works.

```python
# Create a figure of size 6 inches by 4 inches.
plt.figure(figsize=(6,4))

# These two lines of code are used to scale the data points down,
# Or else the data points will be scattered very far apart.

# Create a minimum and maximum range of X1.
x_min, x_max = np.min(X1, axis=0), np.max(X1, axis=0)

# Get the average distance for X1.
X1 = (X1 - x_min) / (x_max - x_min)

# This loop displays all of the datapoints.
for i in range(X1.shape[0]):
    # Replace the data points with their respective cluster value
    # (ex. 0) and is color coded with a colormap (plt.cm.spectral)
    plt.text(X1[i, 0], X1[i, 1], str(y1[i]),
             color=plt.cm.nipy_spectral(agglom.labels_[i] / 10.),
             fontdict={'weight': 'bold', 'size': 9})

# Remove the x ticks, y ticks, x and y axis
plt.xticks([])
plt.yticks([])
#plt.axis('off')



# Display the plot of the original data before clustering
plt.scatter(X1[:, 0], X1[:, 1], marker='.')
# Display the plot
plt.show()
```

```python
dist_matrix = distance_matrix(X1,X1)
print(dist_matrix)
```

```python
Z = hierarchy.linkage(dist_matrix, 'complete')
```

```python
dendro = hierarchy.dendrogram(Z)
```
